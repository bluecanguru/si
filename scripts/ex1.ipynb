Exercise 2 -  NumPy array Indexing/Slicing

2.a) General Examples

```python
import numpy as np
from si.data.dataset import Dataset

X = np.array([
    [1.0, 2.0, np.nan],
    [4.0, np.nan, 6.0],
    [7.0, 8.0, 9.0]
])
y = np.array([0, 1, 1])
features = np.array(['f1', 'f2', 'f3'])
label = "class"

dataset = Dataset(X, y, features, label)

print("\nOriginal dataset:")
print(dataset.X)

#dropna example
dataset_drop = Dataset(X.copy(), y.copy(), features, label).dropna()
print("\nAfter dropna():")
print(dataset_drop.X)

#fillna example
dataset_fill = Dataset(X.copy(), y.copy(), features, label).fillna("mean")
print("\nAfter fillna('mean'):")
print(dataset_fill.X)

#remove_by_index example
dataset_remove = Dataset(X.copy(), y.copy(), features, label).remove_by_index(1)
print("\nAfter remove_by_index(1):")
print(dataset_remove.X)
```

2.b) Iris Dataset Examples:

```python
import numpy as np
from si.data.dataset import Dataset
from si.io.csv_file import read_csv

iris = read_csv("../datasets/iris/iris.csv", label="class")

print("Original Iris dataset shape:", iris.shape())

#dropna example

iris_drop_example = Dataset(iris.X.copy(), iris.y.copy(), iris.features, iris.label)
iris_drop_example.X[0, 1] = np.nan 
iris_drop_example.X[10, 2] = np.nan

print("\nBefore dropna — samples:", iris_drop_example.shape()[0])
iris_drop_example.dropna()
print("After dropna — samples:", iris_drop_example.shape()[0])

#fillna example

iris_fill_example = Dataset(iris.X.copy(), iris.y.copy(), iris.features, iris.label)
iris_fill_example.X[5, 0] = np.nan
iris_fill_example.X[15, 3] = np.nan

print("\nBefore fillna — NaN count:", np.isnan(iris_fill_example.X).sum())
iris_fill_example.fillna("mean")
print("After fillna — NaN count:", np.isnan(iris_fill_example.X).sum())

#remove_by_index example

iris_remove_example = Dataset(iris.X.copy(), iris.y.copy(), iris.features, iris.label)
print("\nBefore remove_by_index — samples:", iris_remove_example.shape()[0])
iris_remove_example.remove_by_index(0) 
print("After remove_by_index — samples:", iris_remove_example.shape()[0])
```

Exercise 7 - KNNRegressor with RMSE

7.3) Test the "KNNRegressor" class using the "cpu.csv" dataset (regression).

```python
from si.io.csv_file import read_csv
from si.data.dataset import Dataset
from si.models.knn_regressor import KNNRegressor
from si.metrics.rmse import rmse
from si.model_selection.split import train_test_split

data = read_csv('../datasets/cpu/cpu.csv', sep=',', features=True, label=True)
dataset = Dataset(data.X, data.y, features=data.features, label=data.label)

train_data, test_data = train_test_split(dataset, test_size=0.2, random_state=42)

knn = KNNRegressor(k=5)
knn.fit(train_data)

y_pred = knn.predict(test_data.X)

rmse_score = rmse(test_data.y, y_pred)
print(f"RMSE: {rmse_score:.1f}")
```

Exercise 8 - RidgeRegression with Least Squares

8.1) Test the "RidgeRegressionLeastSquares" class using the "cpu.csv" dataset.

```python
from si.io.csv_file import read_csv
from si.data.dataset import Dataset
from si.models.ridge_regression_least_squares import RidgeRegressionLeastSquares
from si.model_selection.split import train_test_split

data = read_csv('../datasets/cpu/cpu.csv', sep=',', features=True, label=True)
dataset = Dataset(data.X, data.y, features=data.features, label=data.label)

train_data, test_data = train_test_split(dataset, test_size=0.2, random_state=42)

ridge = RidgeRegressionLeastSquares(l2_penalty=1.0, scale=True)
ridge.fit(train_data)

y_pred = ridge.predict(test_data.X)
mse_score = ridge.score(test_data)
print(f"MSE: {mse_score:.1f}")
```

Exercise 9 - RandomForestClassifier class

9.2) Test the random forest class using the following protocol:
1.Use the iris.csv dataset
2.Split the data into train and test sets
3.Create the RandomForestClassifier model
4.Train the model. What is the score of the model on the test set?

```python
from si.io.csv_file import read_csv
from si.model_selection.split import train_test_split
from si.models.random_forest_classifier import RandomForestClassifier

data = read_csv('../datasets/iris/iris.csv', sep=',', features=True, label=True)

train, test = train_test_split(data, test_size=0.2, random_state=42)

rf = RandomForestClassifier(n_estimators=20, max_depth=5)

rf.fit(train)
score = rf.score(test)
print("Test accuracy:", score)
```
